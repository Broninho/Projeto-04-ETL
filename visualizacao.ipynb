```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# üé® Estilo dos gr√°ficos
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# üìÅ Caminhos dos dados
BASE_DIR = Path().resolve()
GOLD_DIR = BASE_DIR / 'data' / 'gold'

# üóÇÔ∏è Arquivos parquet
gastos_parlamentar = GOLD_DIR / 'gastos_por_parlamentar.parquet'
gastos_categoria = GOLD_DIR / 'gastos_por_categoria_partido.parquet'
top_fornecedores = GOLD_DIR / 'top_fornecedores_partido.parquet'

# ‚úÖ Leitura dos dados
df_parlamentar = pd.read_parquet(gastos_parlamentar)
df_categoria = pd.read_parquet(gastos_categoria)
df_fornecedores = pd.read_parquet(top_fornecedores)

df_parlamentar.head()
```

```python
df_parlamentar['periodo'] = pd.to_datetime(df_parlamentar['ano'].astype(str) + '-' + df_parlamentar['mes'].astype(str))
evolucao = df_parlamentar.groupby('periodo')['valor_documento'].sum().reset_index()

plt.figure()
sns.lineplot(data=evolucao, x='periodo', y='valor_documento')
plt.title('Evolu√ß√£o mensal dos gastos parlamentares')
plt.ylabel('Total gasto (R$)')
plt.xlabel('Per√≠odo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top10 = df_parlamentar.groupby('nome_parlamentar')['valor_documento'].sum().nlargest(10).reset_index()

plt.figure()
sns.barplot(
    data=top10,
    y='nome_parlamentar',
    x='valor_documento',
    hue='nome_parlamentar',
    palette='Blues_d',
    dodge=False,
    legend=False
)
plt.title('Top 10 parlamentares que mais gastaram')
plt.xlabel('Total gasto (R$)')
plt.ylabel('Parlamentar')
plt.tight_layout()
plt.show()
```

```python
categoria_total = df_categoria.groupby('tipo_despesa')['valor_documento'].sum().reset_index().sort_values('valor_documento', ascending=False)

plt.figure()
sns.barplot(data=categoria_total, y='tipo_despesa', x='valor_documento', 
            palette='Greens_d', hue='tipo_despesa', legend=False)
plt.title('Gastos totais por tipo de despesa')
plt.xlabel('Total gasto (R$)')
plt.ylabel('Tipo de despesa')
plt.tight_layout()
plt.show()
```

```python
top10_forn = df_fornecedores.groupby('nome_fornecedor')['valor_documento'].sum().nlargest(10).reset_index()

plt.figure()
sns.barplot(
    data=top10_forn, 
    y='nome_fornecedor', 
    x='valor_documento', 
    palette='Oranges_d', 
    hue='nome_fornecedor', 
    legend=False
)
plt.title('Top 10 fornecedores que mais receberam')
plt.xlabel('Total recebido (R$)')
plt.ylabel('Fornecedor')
plt.tight_layout()
plt.show()
```
